```python
airport_codes = pd.read_csv('https://raw.githubusercontent.com/datasets/airport-codes/master/data/airport-codes.csv')
```

```python
airport_codes = airport_codes[airport_codes['iso_country'] == 'US']
```

```python
airport_codes
```

```python
airport_codes.isnull().count()
```

```python
airport_codes[airport_codes['iata_code'] == 'IAD']
```

```python
airport_codes = airport_codes.drop_duplicates(subset='iata_code', keep="first")
```

```python
airport_codes
```

```python
pd.merge(left = df, right=airport_codes, left_on = "Origin", right_on = "iata_code", how="left")
```

```python
pd.merge(left = df, right=airport_codes[['name', 'municipality', 'iata_code']], left_on = "Origin", right_on = "iata_code", how="left")
```

```python
df = pd.merge(left = df, right=airport_codes[['name', 'municipality', 'iata_code']], left_on = "Origin", right_on = "iata_code", how="left")
```
