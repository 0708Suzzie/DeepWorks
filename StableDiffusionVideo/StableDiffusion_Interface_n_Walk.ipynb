```python
from IPython.display import HTML
from base64 import b64encode
```

```python
from stable_diffusion_videos import StableDiffusionWalkPipeline
from diffusers.schedulers import LMSDiscreteScheduler
import torch
```

```python
pipeline = StableDiffusionWalkPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    use_auth_token=True,
    torch_dtype=torch.float16,
    revision="fp16",
    scheduler=LMSDiscreteScheduler(
        beta_start=0.00085, beta_end=0.012, beta_schedule="scaled_linear", 
    )
).to("cuda")
```

```python
pipeline
```

```python
output_video_path = pipeline.walk(
    ['a wild horse', 'a zebra', 'a donkey', 'a pony', 'a gray horse'],
    [232,5355,3553,5345,2442],
    num_interpolation_steps=5,
    height=512,
    width=512
)
```

```python
def visualize_video_colab(video_path):
    mp4 = open(video_path,'rb').read()
    data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
    return HTML("""
    <video width=400 controls>
        <source src="%s" type="video/mp4">
    </video>
    """ % data_url)
```

```python
visualize_video_colab(output_video_path)
```

```python
# Seconds in the song.
audio_offsets = [10, 12]
fps = 30  # Use lower values for testing (5 or 10), higher values for better quality (30 or 60)

# Convert seconds to frames
num_interpolation_steps = [(b-a) * fps for a, b in zip(audio_offsets, audio_offsets[1:])]
```

```python
num_interpolation_steps
```

```python
output_video_path = pipeline.walk(
    ['a futuristic landscape', 'a relistic landscape'],
    [3553,5345],
    num_interpolation_steps=10, #num_interpolation_steps,
    height=512,
    width=512,
    audio_filepath='/content/guitar.mp3',
    audio_start_sec=audio_offsets[0],
    output_dir='landscape',
    guidance_scale=7.5,    
    num_inference_steps=50
)
```

```python
visualize_video_colab(output_video_path)
```
